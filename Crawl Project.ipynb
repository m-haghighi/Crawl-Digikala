```python
Listgoods = [] 
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
def digikala(goods):
    
    driver=webdriver.Chrome(executable_path=r"E:\ChromeDriver\chromedriver_win32\chromedriver.exe")
    driver.get("https://www.digikala.com/")
    driver.maximize_window()
    search=driver.find_element_by_css_selector("body > header > div > div > div.c-header__right-side > div.c-header__search > div > input")
    search.send_keys(goods)
    search.send_keys(Keys.ENTER)
    
    try:
        for i in range(4):# 4 is number of pages 
            result1 = driver.find_elements_by_css_selector("div.c-product-box__content")
            link = driver.find_element_by_css_selector("div.o-page__content")
            link1=link.find_elements_by_css_selector("a.c-product-box__img.c-promotion-box__image.js-url.js-product-item.js-product-url")
            for i in range(len(result1)):
                Tittle = result1[i].find_element_by_css_selector("div.c-product-box__content--row ").text
                Price = result1[i].find_element_by_css_selector("div.c-product-box__row.c-product-box__row--price").text
                index=Price.rfind("\n")
                if(index!=-1):
                  Price=Price[index+1:]
                Rate = result1[i].find_element_by_css_selector("div.c-product-box__rate-status").text
                index=Rate.find("\n")
                Rate=Rate[0:index]
                Links=link1[i].get_attribute("href")
                Listgoods.append([Tittle,Rate,Price,Links])
        pages = driver.find_elements_by_css_selector("#content > div > div > div.o-page__row.o-page__row--listing > div.o-page__content > div > article > div > div.c-pager > ul > li")[i+1].click()
        time.sleep(30)
    except:
             pass
```

```python
import pandas as pd
from pandas import DataFrame
Listgoods=[]
digikala("گوشی نوکیا")
df = DataFrame(Listgoods,columns=['Tittle','Rate',"Price","links"])
pd.set_option('display.max_rows', df.shape[0]+1)
df
```
